```python
from deepagents import create_deep_agent
from deepagents.backends import StateBackend
```

```python
agent=create_deep_agent(
    model="google_genai:gemini-3.5-flash",
    backend=StateBackend())
```

```python
import importlib.metadata

print(importlib.metadata.version("deepagents"))
print(importlib.metadata.version("langchain-google-genai"))
```

```python
result=agent.invoke({
    "messages":[{
        "role": "user",
        "content": "Create a file /notes/todo.txt wuth exactly the following content:\n\n- Buy milk\n- Call Alice\n- Finish report"
    }]
})
```

```python
result
```

```python
files=result.get("files",{})
if files:
    for path,content in files.items():
        print(f"File: {path}\nContent:\n{content}\n")
else:
    print("No files were created.")
```
