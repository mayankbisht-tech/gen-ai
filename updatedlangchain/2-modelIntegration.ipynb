```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("google_genai:gemini-2.5-flash-lite")
response=model.invoke("hello introuce youself")
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response=model.invoke("hello kemcho")
response.content
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("groq:qwen/qwen3-32b")
response=model.invoke("ask my intro in kumaoni")
response.content
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="qwen/qwen3-32b")
response=model.invoke("Who is bhagat singh")
response.content
```

```python
for chunk in model.stream("write a 200 words paragraph of on ML"):
    print(chunk.text,end=" ",flush=True)
```

```python
#batch
responses=model.batch([
    "who is Bhagat Singh",
    "who is lincon",
    "who is virat kholi"
])
for response in responses:
    print(response)
```

```python
#batch
responses=model.batch([
    "who is Bhagat Singh",
    "who is lincon",
    "who is virat kholi"
],config={
    'max_concurrency':5,
})
for response in responses:
    print(response)
```
