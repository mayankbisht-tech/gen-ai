```python
from pydantic import BaseModel,Field
class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="this is the year at which this movie is released")
    director:str=Field(description="the director of the movie")
    rating:float=Field(description="the movie rating out of 10")
```

```python
from pydantic import BaseModel,Field
class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="this is the year at which this movie is released")
    director:str=Field(description="the director of the movie")
    rating:float=Field(description="the movie rating out of 10")
model_with_struct=model.with_structured_output(Movie,include_raw=True)
model_with_struct.invoke("provide me details about the movie inception")
```

```python
from pydantic import BaseModel,Field
class Actor(BaseModel):
    name:str=Field("Name of the actor")
    role:str=Field("role of the actor in the movie")
class MovieDetails(BaseModel):
    title:str
    year:int
    cast:list[Actor]
    genre:list[str]
    budget:float|None=Field(None,description="budget in million usd")
model_with_struct=model.with_structured_output(MovieDetails)
model_with_struct.invoke("provide me details about the movie inception")
```

```python
#TypeDict
from typing_extensions import TypedDict,Annotated
class MovieDict(TypedDict):
    title:Annotated[str,...,"the title of the movie"]
    year:Annotated[int,...,"the year the movie is released"]
    director:Annotated[str,...,"the director of the moview is "]
    rating:Annotated[float,...,"the movie rating out of 10"]
model_withTypeDict=model.with_structured_output(MovieDict)
model_withTypeDict.invoke("please provide the deatils of the moview avenger")
```
